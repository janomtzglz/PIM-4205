# Sine wave

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

let's create a sine wave

* create an array <br>
* frquecy <br>
* create the plot of the wave

```python
T=np.r_[0:10:0.1]  # shorthand version for T=arange(0,10,0.1)
freq=0.5
X=np.sin(2*np.pi*freq*T) # sine function from numpy
plt.xlabel('Time[sec]') # 
plt.ylabel('Values')
plt.title('Sine wave')
plt.savefig('sinewave.png',pdi=200)
plt.plot(T,X)
```
