# Display of statistical data
### Univariate Data

* import standard packages

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats
```

* generate the data

```python
x=np.random.randn(500)
```

```python
plt.plot(x,'-')
```

```python
plt.plot(x,'.')
```

equivalently: Scatter plot

```python
plt.scatter(np.arange(len(x)),x)
```

### Histograms

Provide a first good overview of the distribution of the data

```python
plt.hist(x, bins=25)
```

### Kernel density estimation
* Using seaborn

Technique to obtain smooth probability densities, i.e. curves describing the likelihood of nding an event in any given interval

```python
sns.kdeplot(x)
```

### Cumulative frequencies
* using scipy

curve indicating the number (or percent) of data with less than a given value

```python
numbins = 25
plt.plot(stats.cumfreq(x,numbins)[0])
```

Display using boxplot
* frequently used in scientific publications to indicate values in two or more groups

```python
plt.boxplot(x, sym='-')
```

### Univariate data grouped bar charts

```python
df = pd.DataFrame(np.random.rand(10, 4),columns=['a', 'b', 'c', 'd'])
df.plot(kind='bar',grid=False)
```
